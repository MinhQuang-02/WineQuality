```python
df.replace({'White Wine': 1, 'Red Wine': 0}, inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

features = df.drop(['quality'], axis=1)
target = df['quality']

# Split the data 80:20 ratio
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=40)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
norm = MinMaxScaler()
X_train_norm = norm.fit_transform(X_train)
X_test_norm = norm.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np

rf = RandomForestClassifier()
rf.fit(X_train_norm, y_train)

# Lấy Importance Score cho mỗi đặc trưng
feature_importances = rf.feature_importances_

# In ra Importance Score của tất cả các đặc trưng
print("Importance Scores for All Features:")
feature_names = ['fixed acidity','volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'Type']
for i, importance_score in enumerate(feature_importances):
    feature_name = feature_names[i]
    print(f"Feature {feature_name}: {importance_score}")

# Sắp xếp các đặc trưng theo giảm dần của Importance Score
sorted_indices = np.argsort(feature_importances)[::-1]

# Chọn số lượng đặc trưng mong muốn (nếu cần)
num_features_to_select = 5
selected_features = sorted_indices[:num_features_to_select]

# In ra các đặc trưng được chọn và Importance Score tương ứng (nếu cần)
print("\nSelected Features:")
for i in selected_features:
    feature_name = feature_names[i]
    importance_score = feature_importances[i]
    print(f"Feature {feature_name}: {importance_score}")

# Xử lý dữ liệu để chỉ giữ lại các đặc trưng được chọn (nếu cần)
X_train_selected = X_train_norm[:, selected_features]
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Tạo một subplot cho mỗi quality
qualities = df['quality'].unique()

# Số lượng đặc trưng bạn muốn hiển thị (tùy chọn)
num_features_to_display = 5

# Vẽ biểu đồ
for quality in qualities:
    plt.figure(figsize=(10, 6))
    plt.suptitle(f"Relationship between Features and Quality {quality}", fontsize=16)

    # Lọc dữ liệu theo quality
    subset = df[df['quality'] == quality]

    # Sắp xếp các đặc trưng theo Importance Score
    sorted_indices = np.argsort(feature_importances)[::-1]
    top_features = sorted_indices[:num_features_to_display]

    # Vẽ biểu đồ cho từng đặc trưng
    for i, feature_index in enumerate(top_features, start=1):
        plt.subplot(2, 3, i)  # Điều chỉnh số hàng và số cột tùy thuộc vào số lượng đặc trưng bạn muốn hiển thị
        plt.scatter(subset[feature_names[feature_index]], subset['quality'], alpha=0.5)
        plt.title(f"Feature {feature_names[feature_index]} vs Quality")
        plt.xlabel(f"Feature {feature_names[feature_index]}")
        plt.ylabel("Quality")

    plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Sử dụng chỉ các đặc trưng quan trọng đã chọn
X_train_selected = X_train_norm[:, selected_features]

# Tạo mô hình RandomForestClassifier mới và huấn luyện với chỉ các đặc trưng quan trọng
rf_selected_features = RandomForestClassifier()
rf_selected_features.fit(X_train_selected, y_train)

# Dự đoán trên tập dữ liệu đào tạo để đánh giá hiệu suất
y_train_pred = rf_selected_features.predict(X_train_selected)

# In độ chính xác trên tập dữ liệu đào tạo
accuracy_train = accuracy_score(y_train, y_train_pred)
print(f"Accuracy on training data: {accuracy_train}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Sử dụng chỉ các đặc trưng quan trọng đã chọn trên tập kiểm thử
X_test_selected = X_test_norm[:, selected_features]

# Dự đoán trên tập kiểm thử
y_test_pred = rf_selected_features.predict(X_test_selected)

# Tính và in confusion matrix
conf_matrix = confusion_matrix(y_test, y_test_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Tính và in classification report
class_report = classification_report(y_test, y_test_pred)
print("Classification Report:")
print(class_report)
```
